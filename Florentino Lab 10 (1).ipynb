```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
def plot_quiv(x,t_mat=np.eye(2)):
    x_prime = x @ t_mat
    size= (2,2)
    plt.figure(figsize=(4,4))

    plt.xlim(-size[0],size[0])
    plt.ylim(-size[1],size[1])
    plt.xticks(np.arange((-size[0]), size[0]+1, 1.0))
    plt.yticks(np.arange((-size[1]), size[1]+1, 1.0))


    plt.quiver([0,0],[0,0], x_prime[0,:], x_prime[1,:], 
               angles='xy', scale_units='xy',scale=1, 
               color=['red','blue'])
    plt.grid()
    plt.show()
```

```python
A = np.array([
    [5, 5],
    [3, 4]
])

plot_quiv(A)
```

```python
t_mat = np.array([
    [-5,6],
    [4,-8]
])
plot_quiv(A)
plot_quiv(A, t_mat)
```

## Shears

```python
D = np.eye(2)

shear1 = np.array([
    [5,6],
    [2,4]
])
shear2 = np.array([
    [1,3],
    [4,-8]
])
plot_quiv(D)
print(D)
plot_quiv(D, shear1)
plot_quiv(D, shear2)
```

```python
shear = np.array([
    [4,7],
    [3,5]
])
plot_quiv(D, shear)
```

```python
shear = np.array([
    [-7,2],
    [3,5]
])
plot_quiv(D, shear)
```

```python
shear = np.array([
    [2.3,7],
    [4.5,3.7]
])
plot_quiv(D, shear)
```

```python
def rot_matrix(theta):
    theta = np.deg2rad(theta)
    rot_mat = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    return rot_mat
```

```python
rotate = rot_matrix(90)
plot_quiv(D)
scale = 4*np.eye(2)
plot_quiv(scale@rotate@D)
```
